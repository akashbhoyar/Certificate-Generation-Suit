```python
import mysql.connector
from tkinter import *
from tkinter import messagebox
from tkinter import ttk
from PIL import Image, ImageDraw, ImageFont
from datetime import date

win=Tk()
win.title("Generating LC")
win.geometry('700x500+290+70')
win.configure(bg='darkseagreen2')

#========Define======#
rn1=StringVar()    # Roll Number
na1=StringVar()    # Name of student
fa1=StringVar()    # father's name 
mo1=StringVar()    # mother's name
ca1=StringVar()    # Caste
pl1=StringVar()    # place of birth
db1=StringVar()    # DOB
db2=StringVar()    # DOB in words
la1=StringVar()    # Last school/college
ad1=StringVar()    # DOA
mb1=StringVar()    # mobile number
p1=StringVar()     # progess
c1=StringVar()     # conduct
dl1=StringVar()    # DOL
ys1=StringVar()    # studing
rl1=StringVar()    # reason of leaving
re1=StringVar()    # remark
lcrl=StringVar()   # roll no for genrating lc

#=======GenerateLC=====#

def generate():
    global rn1
    global p1
    global c1
    global dl1
    global ys1
    global rl1
    global re1
    
    roll_no=rn1.get()
    progress=p1.get()
    conduct=c1.get()
    date_leaving=dl1.get()
    year_studying=ys1.get()
    reason_leaving=rl1.get()
    remark=re1.get()
    
    if (roll_no=="" or progress=="" or conduct=="" or date_leaving=="" or year_studying=="" or reason_leaving=="" or remark==""):
        messagebox.showinfo("Shudent Info","All fields are required to fill")
    
    else:
        conn=mysql.connector.connect(host='localhost',user='root',passwd='********',database="project")
        query2="update prot2 set roll_no='"+roll_no+"', progress='"+progress+"',conduct='"+conduct+"',date_leaving='"+date_leaving+"',year_studying='"+year_studying+"',reason_leaving='"+reason_leaving+"',remark='"+remark+"' where roll_no='"+roll_no+"'"
        mycur=conn.cursor()
        mycur.execute(query2)
        conn.commit()
        conn.close()
        messagebox.showinfo("Student Info ","Student info Submitted")
        rn1.set('')
        p1.set('')
        c1.set('')
        dl1.set('')
        ys1.set('')
        rl1.set('')
        re1.set('')
        

# def printlc():
#     q=win.textarea.get('1.0',END)
#     if q!='':
#         filename='C:\\Users\\acer\\PycharmProjects\\python Project in GUI\\Student Record\\'+(_____.get())+'.png'
#         open(filename,'w').write(q)
#         os.startfile(filename,'print')
#     else:
#         messagebox.showerror('Error','There is nothing to be print')
        

#=======================LC===========================================#
def savelc():
    global lcrl
    lcrl1=lcrl.get()
    if lcrl1=="":
        messagebox.showinfo("Student Info ","All fields are required to fill")
    else:
        conn=mysql.connector.connect(host='localhost',user='root',passwd='********',database="project")
        qur1="SELECT * FROM prot2 WHERE roll_no='%s'"%(lcrl1)
        mycur=conn.cursor()
        mycur.execute(qur1)
        result=mycur.fetchall()
        for i in result:
            a=list(i)
        rn2=str(a[0])    # Roll Number
        na2=str(a[1])    # Name of student
        fa2=str(a[2])    # father's name 
        mo2=str(a[3])    # mother's name
        ca2=str(a[4])    # Caste
        Nation="Indian"
        pl2=str(a[5])    # place of birth
        dbi1=str(a[6])   # DOB
        dbw2=str(a[7])   # DOB in words
        la2=str(a[8])    # Last school/college
        ad2=str(a[9])    # DOA
        mb2=str(a[10])    # mobile number
        p2=str(a[11])     # progess
        c2=str(a[12])     # conduct
        dl2=str(a[13])    # DOL
        ys2=str(a[14])    # studing
        rl2=str(a[15])    # reason of leaving
        re2=str(a[16])    # remark
        today = str(date.today())
        

          
        my_image = Image.open("lc.jpg")
        title_font = ImageFont.truetype('arial.ttf', 25)
        date_font  = ImageFont.truetype('arial.ttf', 18)
        image_editable = ImageDraw.Draw(my_image)
        
        image_editable.text((500,490), rn2, (0, 0, 0), font=title_font)
        image_editable.text((910,630), na2, (0, 0, 0), font=title_font)
        image_editable.text((910,680), fa2, (0, 0, 0), font=title_font)
        image_editable.text((910,730), mo2, (0, 0, 0), font=title_font)
        image_editable.text((910,820), ca2, (0, 0, 0), font=title_font)
        image_editable.text((910,925), Nation, (0, 0, 0), font=title_font)
        image_editable.text((910,975), pl2, (0, 0, 0), font=title_font)
        image_editable.text((910,1040), dbi1, (0, 0, 0), font=title_font)
        image_editable.text((910,1090), dbw2, (0, 0, 0), font=title_font)
        image_editable.text((910,1165), la2, (0, 0, 0), font=title_font)
        image_editable.text((910,1215), ad2, (0, 0, 0), font=title_font)
        image_editable.text((910,1265), p2, (0, 0, 0), font=title_font)
        image_editable.text((910,1315), c2, (0, 0, 0), font=title_font)
        image_editable.text((910,1360), dl2, (0, 0, 0), font=title_font)
        image_editable.text((910,1410), ys2, (0, 0, 0), font=title_font)
        image_editable.text((910,1460), rl2, (0, 0, 0), font=title_font)
        image_editable.text((910,1510), re2, (0, 0, 0), font=title_font)
        image_editable.text((165,1660), today, (0, 0, 0), font=date_font)
        
        my_image.save(rf'{na2}.jpg')
#         my_image.show()
            
    
#=======Lables=======#

#Roll Number
progresslabel=Label(win,text="Roll No",bg='white',bd=5,width=20,fg='black',relief='ridge',font=('times new roman',12,'bold'))    #win is object
progresslabel.place(x=30,y=30)
progressent=Entry(win,bd=5,fg='black',bg='white',width=45,textvariable=rn1,font=('times new roman',12,'bold'))
progressent.place(x=250,y=30)

                  
#Progress
progresslabel=Label(win,text="Progress",bg='white',bd=5,width=20,fg='black',relief='ridge',font=('times new roman',12,'bold'))    #win is object
progresslabel.place(x=30,y=70)
progressent=Entry(win,bd=5,fg='black',bg='white',width=45,textvariable=p1,font=('times new roman',12,'bold'))
progressent.place(x=250,y=70)

#Conduct

conductlabel=Label(win,text="Conduct",bg='white',bd=5,width=20,fg='black',relief='ridge',font=('times new roman',12,'bold'))    #win is object
conductlabel.place(x=30,y=110)
conductent=Entry(win,bd=5,fg='black',bg='white',width=45,textvariable=c1,font=('times new roman',12,'bold'))
conductent.place(x=250,y=110)

#Date of leaving
DOLlabel=Label(win,text="Date of Leaving",bg='white',bd=5,width=20,fg='black',relief='ridge',font=('times new roman',12,'bold'))    #win is object
DOLlabel.place(x=30,y=150)
DOLent=Entry(win,bd=5,fg='black',bg='white',width=45,textvariable=dl1,font=('times new roman',12,'bold'))
DOLent.place(x=250,y=150)

#He was Studying in
yslabel=Label(win,text="He was Studying in",bg='white',bd=5,width=20,fg='black',relief='ridge',font=('times new roman',12,'bold'))    #win is object
yslabel.place(x=30,y=190)
ysent=Entry(win,bd=5,fg='black',bg='white',width=45,textvariable=ys1,font=('times new roman',12,'bold'))
ysent.place(x=250,y=190)

#Reason
reasonlabel=Label(win,text="Reason for Leaving",bg='white',bd=5,width=20,fg='black',relief='ridge',font=('times new roman',12,'bold'))    #win is object
reasonlabel.place(x=30,y=230)
reasonent=Entry(win,bd=5,fg='black',bg='white',width=45,textvariable=rl1,font=('times new roman',12,'bold'))
reasonent.place(x=250,y=230)

#Remark
remarklabel=Label(win,text="Remark",bg='white',bd=5,width=20,fg='black',relief='ridge',font=('times new roman',12,'bold'))    #win is object
remarklabel.place(x=30,y=270)
remarkent=Entry(win,bd=5,fg='black',bg='white',width=45,textvariable=re1,font=('times new roman',12,'bold'))
remarkent.place(x=250,y=270)


#=======Buttons======#

generatebtn=Button(win,text='Submit',width=20,relief='ridge',command=generate).place(x=250,y=330)

#===========For save LC========# 
remarklabel=Label(win,text="Save LC by Roll No.",bg='white',bd=5,width=20,fg='black',relief='ridge',font=('times new roman',12,'bold'))    #win is object
remarklabel.place(x=30,y=390)

remarkent=Entry(win,bd=5,fg='black',bg='white',width=20,textvariable=lcrl,font=('times new roman',12,'bold'))
remarkent.place(x=250,y=390)

printbtn=Button(win,text='Save LC',width=20,relief='ridge',command=savelc).place(x=450,y=390)


win.mainloop()
```
